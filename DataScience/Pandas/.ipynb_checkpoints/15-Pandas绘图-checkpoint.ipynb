```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
s = pd.Series(np.random.randn(10),index=np.arange(0,100,10))
print(s.plot())
```

```python
df = pd.DataFrame(np.random.randn(10,4).cumsum(0),
                  index=np.arange(0,100,10),
                  columns=['A','B','C','D'])
print(df.head())
```

```python
print(df.plot())
```

```python
data = pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
print(data.head())
```

axes[0]表示index水平展示 axes[1]表示index垂直展示 kind=bar柱状图 kind=barh柱状图水平展示


```python
import matplotlib.pyplot as plt
```

```python
fig,axes= plt.subplots(2,1)
print(data.plot(ax=axes[0],kind='bar'))
print(data.plot(ax=axes[1],kind='barh'))
```
