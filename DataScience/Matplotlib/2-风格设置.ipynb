```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
print(plt.style.available)
```

```python
x = np.linspace(-10,10)
y = np.sin(x)
```

```python
plt.style.use('dark_background')
plt.plot(x,y)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.plot(x,y)
plt.show()
```

```python
plt.style.use('seaborn')
plt.plot(x,y)
plt.show()
```

```python
plt.style.use(['seaborn','bmh'])
plt.plot(x,y)
plt.show()
```

```python
plt.xkcd()
plt.plot(x,y)
plt.show()
```
