```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
height=[161,170,182,175,173,165]
weight=[50,58,80,70,69,55]
plt.scatter(height,weight)
plt.show()
```

```python
N=1000
x=np.random.randn(N)
y1=np.random.randn(len(x))
plt.scatter(x,y1)
plt.show()
```

```python
y2=x+np.random.randn(len(x))*0.1
plt.scatter(x,y2)
plt.show()
```

```python

y3=-1*x+np.random.randn(len(x))*0.1
plt.scatter(x,y3)
plt.show()
```

```python
op,close=np.loadtxt('data/000001.csv',delimiter=',',skiprows=1,usecols=(1,4),unpack=True)
change=close-op
yesterday=change[:-1]
today=change[1:]
plt.scatter(yesterday,today)
plt.show()
```

```python
# 颜色, c
# 点大小, s
# 透明度, alpha
# 点形状, marker
plt.scatter(yesterday,today,s=50,c='r',marker='<',alpha=0.5)
plt.show()
```
