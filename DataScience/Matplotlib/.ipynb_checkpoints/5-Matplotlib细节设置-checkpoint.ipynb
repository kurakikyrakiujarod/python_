```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

### 设置不显示x轴和y轴的刻度

```python
x = range(10)
y = range(10)
plt.figure(figsize=(20,12),dpi=80)
fig = plt.gca()
plt.plot(x,y)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.show()
```

### 设置不显示图表的上边和右边的边框线条

```python
import math
plt.figure(figsize=(20,12),dpi=80)
x = np.random.normal(loc=0.0,scale=1.0,size=300)
width= 0.5
# print(x)
# print(math.floor(x.min()))
# print(math.ceil(x.max()))
bins = np.arange(math.floor(x.min()-width),math.ceil(x.max())+width,width)
# print(bins)
ax = plt.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['left'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# plt.tick_params(bottom=False,top=False,left=False,right=False,colors='red')
plt.grid()
plt.hist(x,alpha=0.5,bins=bins)
plt.show()
```

### 设置X轴标题的属性

```python
x = range(10)
y = range(10)
labels = ['python' for i in range(10)]
plt.figure(figsize=(20,12),dpi=80)
fig,ax = plt.subplots()
plt.plot(x,y)
ax.set_xticklabels(labels=labels,rotation=90,horizontalalignment='left')
plt.show()
```

### 设置图例

```python
x = np.arange(10)
plt.figure(figsize=(20,12),dpi=80)
for i in range(1,4):
    plt.plot(x,i*x**2,label='group %d'%i)
plt.legend(loc='best')
plt.show()
```

### 图例属性详细设置

```python
fig = plt.figure(figsize=(20,12),dpi=80)
ax = plt.subplot(111)
x = np.arange(10)
for i in range(1,4):
    plt.plot(x,i*x**2,label='group %d'%i)
ax.legend(loc='center',bbox_to_anchor=(0.2,0.5),ncol=3)
plt.show()
```
