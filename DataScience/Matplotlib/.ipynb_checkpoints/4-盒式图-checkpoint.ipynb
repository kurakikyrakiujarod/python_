```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 绘制单个盒式图

```python
np.random.seed(100)
data = np.random.normal(size=(1000,),loc=0,scale=1)
plt.figure(figsize=(20,8),dpi=80)
plt.boxplot(data,sym='o',whis=1.5)
plt.show()
```

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,5)]
fig = plt.figure(figsize=(8,6),dpi=80)
plt.boxplot(tang_data,notch=False,sym='s',vert=True)
plt.xticks([y+1 for y in range(len(tang_data))],['x%d'%n for n in range(1,5)])
plt.xlabel('x')
plt.title('box plot')
plt.show()
```

### 改变盒式图线条颜色

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,5)]
fig = plt.figure(figsize=(8,6),dpi=80)
bplot = plt.boxplot(tang_data,notch=False,sym='s',vert=True)
plt.xticks([y+1 for y in range(len(tang_data))],['x%d'%n for n in range(1,5)])
plt.xlabel('x')
plt.title('box plot')
for box in bplot.keys():
    for line in bplot[box]:
        line.set_color('black')
plt.show()
```

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,5)]
fig = plt.figure(figsize=(8,6),dpi=80)
bplot = plt.boxplot(tang_data,notch=True,sym='s',vert=False)
plt.yticks([y+1 for y in range(len(tang_data))],['x%d'%n for n in range(1,5)])
plt.ylabel('x')
plt.title('box plot')

for box in bplot.keys():
    for line in bplot[box]:
        line.set_color('black')
plt.show()
```

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6),dpi=80)
bplot = plt.boxplot(tang_data,notch=True,sym='s',vert=False,patch_artist=True)
plt.yticks([y+1 for y in range(len(tang_data))],['x%d'%n for n in range(1,5)])
plt.ylabel('x')
plt.title('box plot')

colors = ['pink','lightblue','lightgreen']
for pathch,color in zip(bplot['boxes'],colors):
    pathch.set_facecolor(color)
plt.show()
```

### 小提琴图

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
tang_data = [np.random.normal(0,std,100) for std in range(6,10)]
axes[0].violinplot(tang_data,showmeans=False,showmedians=True)
axes[0].set_title('violin plot')

axes[1].boxplot(tang_data)
axes[1].set_title('box plot')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(tang_data))])
plt.setp(axes,xticks=[y+1 for y in range(len(tang_data))],xticklabels=['x%d'%n for n in range(1,5)])
plt.show()
```
