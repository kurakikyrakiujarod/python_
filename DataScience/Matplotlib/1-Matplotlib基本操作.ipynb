# Matplotlib基本操作

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.xlabel('xlable',fontsize=16)
plt.ylabel('ylable',fontsize=16)
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

### 线条

```python
# '.'       point marker
# ','       pixel marker
# 'o'       circle marker
# 'v'       triangle_down marker
# '^'       triangle_up marker
# '<'       triangle_left marker
# '>'       triangle_right marker
# '1'       tri_down marker
# '2'       tri_up marker
# '3'       tri_left marker
# '4'       tri_right marker
# 's'       square marker
# 'p'       pentagon marker
# '*'       star marker
# 'h'       hexagon1 marker
# 'H'       hexagon2 marker
# '+'       plus marker
# 'x'       x marker
# 'D'       diamond marker
# 'd'       thin_diamond marker
# '|'       vline marker
# '_'       hline marker
```

```python
plt.xlabel('xlable',fontsize=16)
plt.ylabel('ylable',fontsize=16)
plt.plot([1,2,3,4,5],[1,4,9,16,25],'x')
plt.show()
```

### 颜色

```python
# b: blue
# g: green
# r: red
# c: cyan
# m: magenta
# y: yellow
# k:  black
# w: white
```

```python
plt.xlabel('xlable',fontsize=16)
plt.ylabel('ylable',fontsize=16)
plt.plot([1,2,3,4,5],[1,4,9,16,25],'--',color='c')
plt.show()
```

```python
plt.xlabel('xlable',fontsize=16)
plt.ylabel('ylable',fontsize=16)
plt.plot([1,2,3,4,5],[1,4,9,16,25],'r--')
plt.show()
```

```python
tang_numpy = np.arange(0,10,0.5)
plt.plot(tang_numpy,tang_numpy,'r--',
         tang_numpy,tang_numpy**2,'bs',
         tang_numpy,tang_numpy**3,'go')
plt.show()
```

```python
# 指定线条的宽度
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y,linewidth=2.0)
plt.show()
```

```python
plt.plot(x,y,color='b',linestyle=':',marker='o',markerfacecolor='r',markersize=10)
plt.show()
```

```python
line = plt.plot(x,y)
plt.setp(line,color='r',linewidth=2.0,alpha=0.4)
plt.show()
```

### 子图

```python
#  211表示一会要画的图是2行一列的 最后一个1表示的是子图当中的第1个图
#  212表示一会要画的图是2行一列的 最后一个2表示的是子图当中的第2个图
plt.subplot(211)
plt.plot(x,y,color='r')

plt.subplot(212)
plt.plot(x,y,color='b')

plt.show()
```

```python
plt.subplot(321)
plt.plot(x,y,color='r')

plt.subplot(324)
plt.plot(x,y,color='b')

plt.show()
```

### 给图上加上注释

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('tang yu di')
plt.text(0,0,'tang yu di')
plt.grid(True)
plt.annotate('tang yu di',xy=(-5,0),xytext=(-2,0.3),
             arrowprops=dict(facecolor='black',shrink=0.05,
             headwidth=10,headlength=10))
plt.plot(x,y,color='b',linestyle=':',marker='o',markerfacecolor='r',markersize=10)
plt.show()
```
